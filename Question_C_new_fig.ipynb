```python
import numpy as np
import os
import matplotlib.pyplot as plt
from src.Question_C_new_funcs import monte_carlo_dla_ps

np.random.seed(10)
save_folder = 'fig'
os.makedirs(save_folder, exist_ok=True)

list_ps = [0.1, 0.4, 0.8]

# Create a single figure with three subplots
fig, axes = plt.subplots(1, 3, figsize=(18, 6))  # 1 row, 3 columns

fig.suptitle("Monte Carlo DLA Growth", fontsize=18, fontweight="bold")

for i, ps in enumerate(list_ps):
    dla_grid_mc = monte_carlo_dla_ps(grid_size=(100, 100), steps=10000000, ps=ps)
    
    axes[i].imshow(dla_grid_mc, origin="upper", cmap="plasma")
    axes[i].set_title(f"ps = {ps}", fontsize = 16)
    axes[i].axis("off")
    
save_path = os.path.join(save_folder, f'Question C.png')
plt.savefig(save_path, dpi=300)
plt.show()

```
