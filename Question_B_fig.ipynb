```python
import numpy as np
import os
import matplotlib.pyplot as plt
from src.Question_B_funcs import monte_carlo_dla

save_folder = 'fig'
os.makedirs(save_folder, exist_ok=True)
dla_grid_mc = monte_carlo_dla(grid_size=(100, 100), steps=100000000)

plt.figure(figsize=(10, 10))
plt.imshow(dla_grid_mc, origin="lower", cmap="plasma")
plt.title("Monte Carlo DLA Growth")
plt.axis("off")
save_path = os.path.join(save_folder, f'Question B.png')
plt.savefig(save_path, dpi=300)
plt.show()

# Considering that a random walker can only start from a random spot of the top boundary and moving up above the boundary deletes
# the random walker, it is very difficult for the cluster to develop throughout the grid, especially close to the 
# horizontal boundaries. On the other hand, for Laplace DLA the cluster is spread throughout the grid. This is the case
# since lower concentration areas are favoured for growth.
```
